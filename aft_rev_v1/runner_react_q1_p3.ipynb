```python
from openai import OpenAI
from pydantic import BaseModel
from typing import Dict, List,Tuple,Optional
from dotenv import load_dotenv
from utils import *
from chatty_boi import *
import os
from tqdm import tqdm
```

```python
with open("./prompt_lib/base_prompt.txt", "r") as f:
    prompt = f.read()

with open("./prompt_lib/constraint_prompt.txt", "r") as f:
    constraints_prompt = f.read()

with open("./prompt_lib/optimize_prompt.txt", "r") as f:
    optimize_prompt = f.read()

input_data = {}

with open("./prompt_lib/base_inputs.txt", "r") as f:
    exec(f.read(),{}, input_data)


base_prompt = prompt
constraints_prompt = constraints_prompt
optimize_prompt = optimize_prompt

node_dict_input = input_data["node_dict_input"]
example_members_input = input_data["example_members_input"]
area_id = input_data["area_id"]
load = input_data["load"]
supports = input_data["supports"]

max_stress = 30
max_mass = 30
max_attempts = 10
runs = 10

folder_name = "q1p3_s30W30_mini_t12"
model = "gpt-4.1-mini"
```

```python
for j in tqdm(range(0,runs)): 
    msg = base_prompt.format(node_dict=node_dict_input, load =load, supports=supports, area_id=area_id, example_member = example_members_input, max_stress= max_stress, max_weight= max_mass)
    max_iterations = 25

    temp_history = {} 

    for i in range(max_iterations):
        print(f"\n\n RUN: {j} | Iteration {i} \n")

        response_, t_gen, attempts, node_dict_g, member_dict_g, res, result_dict,stop = get_answer(model=model, max_attempts = 10, message=msg, broken_folder = f"./results_final_runs/broken_results_q1/{folder_name}/run_{j}/", identifier = f"step_{i}", response_folder = f"./results_final_runs/responses_q1/{folder_name}/run_{j}/", load = load, supports = supports, init_node_dict = node_dict_input, stw_ = True)

        result_dict_cloned = result_dict.copy()

        temp_history[f"iteration_{i}"] = {
            "node_dict": node_dict_g,
            "member_dict": member_dict_g,
            "result_dict": result_dict,
            "iteration": i  # Keep original index for constraint_not_met naming
        }

        # === LIVE RANKED HISTORY FOR PROMPT ===
        ranked_history = {}

        # Collect feasible entries
        feasible_entries = [
            (key, val["result_dict"]["max_stress"])
            for key, val in temp_history.items()
            if val["result_dict"]["max_stress"] <= max_stress and
            val["result_dict"]["total_mass"] <= max_mass
        ]

        # Sort by STW (score)
        feasible_entries.sort(key=lambda x: x[1])

        # Add feasible entries
        for rank, (key, _) in enumerate(feasible_entries, start=1):
            val = temp_history[key]
            result_dict = val["result_dict"]
            ranked_history[f"rank_{rank}"] = {
                "rank": rank,
                "score": round(result_dict["max_stress"], 4),  # 
                "structure": {
                    "node_dict": val["node_dict"],
                    "member_dict": val["member_dict"]
                },
                "fem_result": result_dict
            }

        # Collect infeasible entries
        infeasible_entries = [
            (key, val["result_dict"]["max_stress"])
            for key, val in temp_history.items()
            if not (
                val["result_dict"]["max_stress"] <= max_stress and
                val["result_dict"]["total_mass"] <= max_mass
            )
        ]

        # Sort infeasible entries by best (lowest) STW
        infeasible_entries.sort(key=lambda x: x[1])

        # Add infeasible entries as constraint_not_met_{rank}
        for rank, (key, score) in enumerate(infeasible_entries, start=1):
            val = temp_history[key]
            result_dict = val["result_dict"]
            ranked_history[f"constraint_not_met_{rank}"] = {
                "rank": "constraint not met",
                "score": round(score, 4),
                "structure": {
                    "node_dict": val["node_dict"],
                    "member_dict": val["member_dict"]
                },
                "fem_result": result_dict
            }



        # Assign ranked version as final history
        history = ranked_history

        
        #save history dict as txt
        #check if folder exists
        if not os.path.exists(f"./results_final_runs/history_q1/{folder_name}/run_{j}/"):
            os.makedirs(f"./results_final_runs/history_q1/{folder_name}/run_{j}")

        with open(f"./results_final_runs/history_q1/{folder_name}/run_{j}/history.txt", "w") as f:
            for key, val in history.items():
                f.write(f"{key}: {val}\n")
        #save response as txt

        if result_dict_cloned["max_stress"] <= max_stress and result_dict_cloned["total_mass"] <= max_mass:
            print("\n Optimizing the design.")
            
            msg = optimize_prompt.format(load =load, supports=supports, area_id=area_id, example_member = example_members_input, max_stress= max_stress, max_weight= max_mass, node_dict_c = node_dict_g, member_dict_c = member_dict_g, fem_result = result_dict_cloned, history = history,iteration=str(i), c_max_stress = result_dict_cloned["max_stress"], total_mass = result_dict_cloned["total_mass"])
            if stop == "True":
                print("\n Stopping the optimization process.")
                break

        else:
            print("\n Constraints not met.")
            
            msg = constraints_prompt.format(load =load, supports=supports, area_id=area_id, example_member = example_members_input, max_stress= max_stress, max_weight= max_mass, node_dict_c = node_dict_g, member_dict_c = member_dict_g, fem_result = result_dict_cloned, history = history, iteration=str(i), c_max_stress = result_dict_cloned["max_stress"], total_mass = result_dict_cloned["total_mass"])
```
